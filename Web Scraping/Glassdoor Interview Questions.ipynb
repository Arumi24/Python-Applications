```python
from bs4 import BeautifulSoup
import requests
import numpy as np
```

```python

list=np.arange(1,13)
count=1

file = open("Risk_Analyst.html","w") 

file.write("<br />")

file.write("<h1>Risk Analyst Interview Questions</h1>")

file.write("<br />")

for i in list:
    print(i)
    string="_IP"
    
    if i==1:
        url="https://www.glassdoor.ca/Interview/risk-analyst-interview-questions-SRCH_KO0,12.htm"
        html=requests.get(url,headers={'User-Agent': 'Mozilla/5.0'})
        content=html.content
        soup=BeautifulSoup(content,"html.parser")

        all=soup.find_all("div",{"class":"interviewQuestionWrapper"})


        for j in range(len(all)):
            
            file.write(str(count)+"<br /> ")
            print(all[j].find("span",{"class":"authorInfo"}).text)
            file.write("<u>"+all[j].find("span",{"class":"authorInfo"}).text+"</u>"+"<br />")
            print(all[j].find("p",{"class":"questionText h3"}).text)
            file.write("<b>Question: </b>"+all[j].find("p",{"class":"questionText h3"}).text+"<br />")
            
            if all[j].find("p",{"class":"cell noMargVert padTop tightBot"})!=None:
                print(all[j].find("p",{"class":"cell noMargVert padTop tightBot"}).text)
                file.write("<b>Answered: </b>"+all[j].find("p",{"class":"cell noMargVert padTop tightBot"}).text+"<br />")
            
            
            file.write("<br />")
            file.write("<br />")
            file.write("<br />")
            print("")
            count=count+1
            
    else:
        
        string=string+str(i)
        url="https://www.glassdoor.ca/Interview/risk-analyst-interview-questions-SRCH_KO0,12"+string+".htm"
        html=requests.get(url,headers={'User-Agent': 'Mozilla/5.0'})
        content=html.content
        soup=BeautifulSoup(content,"html.parser")

        all=soup.find_all("div",{"class":"interviewQuestionWrapper"})


        for j in range(len(all)):
            
            file.write(str(count)+"<br /> ")
            
            print(all[j].find("span",{"class":"authorInfo"}).text)
            file.write("<u>"+all[j].find("span",{"class":"authorInfo"}).text+"</u>"+"<br />")
            print(all[j].find("p",{"class":"questionText h3"}).text)
            file.write("<b>Question: </b>"+all[j].find("p",{"class":"questionText h3"}).text.encode('utf-8')+"<br />")
            
            
            if all[j].find("p",{"class":"cell noMargVert padTop tightBot"})!=None:
                print(all[j].find("p",{"class":"cell noMargVert padTop tightBot"}).text)
                file.write("<b>Answered: </b>"+all[j].find("p",{"class":"cell noMargVert padTop tightBot"}).text.encode('utf-8')+"<br />")
            
            file.write("<br />")
            file.write("<br />")
            file.write("<br />")
            print("")
            count=count+1
        
file.close()
print("done")
print(count)
```
